```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_csv('../data/merged_oil_wells.csv', low_memory=False)
df.head()
```

```python
df.info()
```

```python
# Identify where the values in two columns differ
col1 = 'Surface_Longitude'
col2 = 'Bottom_hole_longitude'

# Create a boolean mask where the values differ
difference_mask = df[col1] != df[col2]

# Get the indices and the corresponding values where they differ
differences = df[difference_mask][[col1, col2]]

# Print the differences
print(f"Differences between {col1} and {col2}:")
print(differences)
```
