```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
file_paths = glob.glob("data/oil_production/Prod*.csv")
data_frames = []
for file_path in file_paths:
    df = pd.read_csv(file_path)
    data_frames.append(df)

combined_production_data = pd.concat(data_frames, ignore_index=True)
combined_production_data
```

```python
oil_wells_data = pd.read_csv('data/oil_wells.csv', low_memory=False)
merged_data = pd.merge(oil_wells_data, combined_production_data, on='API_WellNo', how='outer')
merged_data.to_csv('data/merged_oil_wells.csv', index=False)
merged_data
```
